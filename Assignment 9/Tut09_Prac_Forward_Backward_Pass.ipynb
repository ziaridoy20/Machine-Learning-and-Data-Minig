```python
import numpy as np
```

### define sigmoid function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
i = [0.3, 0.4] # inputs 
w12 = [0.5, 0.1] # weight1, weight2
b1 = 0.33 #first bias

w3 = 0.85 # weight3
b2 = 0.1 # Second bias

w4 = 0.45 # weight4
```

### Input of first hidden layer

```python
# Now calculate to input of first hidden layer

h1_input = np.dot(i,w12) + b1
print (h1_input)
```

### Output of first hidden layer

```python
# Apply sigmoid function on input of h1 to calculate the output of first hidden layer

h1_output = sigmoid(h1_input)

print(h1_output)
```

### Input of Second hidden layer

```python
# That is:
#    ( (output of h1) * (Weight3)) + Second bias
h2_input = (h1_output * w3)  + b2

print(h2_input)
```

### Output of second hidden layer

```python
# apply sigmoid function to input of h2

h2_output = sigmoid(h2_input)

print(h2_output)
```

### Input of last (output) layer

```python
outLayer_input = h2_output * w4

print(outLayer_input)
```

### Output of last (output) layer

```python
# apply sigmoid function to input of last layer

outLayer_output = sigmoid(outLayer_input)

print(outLayer_output)
```

### Total Error in output

```python
desired_output = 0.01
```

```python
total_error = (1 / 2) * ((outLayer_output - desired_output) **2)

print(total_error)
```
